```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set the parameters 
num_samples = 1000
sample_size = 30
distribution_range = (0,1)

# genterate samples from uniform distribution 
sample = np.random.uniform(distribution_range[0],distribution_range[1],(num_samples,sample_size))

# calculate the mean of each sample 
sample_mean = np.mean(sample,axis=1)
# plot the histogram of sample means
plt.hist(sample_mean,bins=30)
plt.show()
```

# Proving that the if the mean of the population is m and variance is s^2 then the mean for the normal distribution will be m and variance will be s^2/n where n is the number of samples

```python
# set the parameters 
num_samples = 10000
sample_size = 50

#gamma distribution parameters
gamma_shape = 2
gamma_scale = 1

#calculate the theoritical mean and variance
theoritical_mean = gamma_shape * gamma_scale
theoritical_variance = gamma_shape * (gamma_scale**2)

# generate the samples from gamma distribution
sample1 = np.random.gamma(shape=gamma_shape,scale=gamma_scale, size=(num_samples, sample_size))

# calculate the mean and variance of each sample
sample_mean1 = np.mean(sample1, axis=1)



# plot the histogram of sample means
plt.hist(sample_mean, bins=30)
plt.show()
```

```python
#calculate the empirical mean and variance
empirical_mean = np.mean(sample_mean1)
empirical_variance = np.var(sample_mean1)

# print the results
print("Theoritical Mean:", theoritical_mean)
print("Theoritical Variance:", theoritical_variance)
print("Empirical Mean:", empirical_mean)
print("Empirical Variance:", empirical_variance)
```
