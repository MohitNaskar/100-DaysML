```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

def find_project_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p / ".git").exists() or (p / "Datasets").exists():
            return p
    raise FileNotFoundError(f"Project root not found from: {start}")

project_root = find_project_root(Path.cwd())
train_path = project_root / "DataAnalysis" / "stats" / "train.csv"
test_path = project_root / "DataAnalysis" / "stats" / "test.csv"

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
final = pd.concat([train, test], ignore_index=True)
```

```python
final.sample(5)
```

```python
pop = final['Age'].dropna()
sample_age = pop.sample(25, random_state=42)
```

# H0 or mean age is 35 for null hypothesis
# H1 or mean age is less than 35 for alternate hypothesis

```python
from scipy.stats import shapiro
stat, p = shapiro(sample_age)
```

```python
print(p)
```

```python
from scipy import stats
t_statistic, p_value = stats.ttest_1samp(sample_age, pop.mean())
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

## Independent Two-Sample t-Test: Problem Statement and Data Design

### Problem Statement
A school wants to compare exam performance between two independent groups of students who used different preparation methods:

- Group A: Traditional classroom revision
- Group B: App-based adaptive practice

We want to test whether the mean exam scores are different between these two groups.

- Null hypothesis ($H_0$): $\mu_A = \mu_B$
- Alternative hypothesis ($H_1$): $\mu_A \ne \mu_B$
- Significance level: $\alpha = 0.05$

### Data Design (Independent Samples)
- Outcome variable: Exam score (0 to 100)
- Group variable: Preparation method (A or B)
- Each student appears in only one group (independent observations)
- Sample size: 15 students per group

Group A scores:
[72, 68, 75, 70, 74, 69, 71, 73, 67, 76, 72, 70, 74, 71, 69]

Group B scores:
[78, 82, 80, 79, 81, 77, 83, 80, 78, 84, 79, 82, 81, 80, 77]

These two samples can be used directly in an independent two-sample t-test.

## Independent Two-Sample t-Test: Problem Statement

A school wants to compare math performance between students taught with two different methods.

- Group A: Traditional teaching method
- Group B: New interactive teaching method
- Outcome: Final exam score (out of 100)

### Hypotheses
- H0: The mean exam score of Group A equals the mean exam score of Group B.
- H1: The mean exam score of Group A is different from the mean exam score of Group B.

This is an independent t-test because the students in Group A and Group B are different individuals (independent groups).

```python
import pandas as pd
import numpy as np

# Designed sample data for two independent groups
group_a_scores = np.array([68, 72, 75, 70, 74, 69, 71, 73, 76, 72, 70, 74])
group_b_scores = np.array([74, 78, 80, 77, 79, 76, 81, 75, 78, 80, 77, 79])

#perform levene's test for equal variances
from scipy.stats import levene
stat, p = levene(group_a_scores, group_b_scores)
print(f"Levene's test statistic: {stat}, P-value: {p}")
```
