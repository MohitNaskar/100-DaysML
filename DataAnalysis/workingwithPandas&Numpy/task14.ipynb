```python
import numpy as np
```

```python
# Find the nearest element in the array to a given integer
a=23 
array = np.array([10, 17, 24, 31, 38, 45, 52, 59])

index = (np.abs(array - a)).argmin()
print(array[index])
```

```python
# Replace mutiples of 3 or 5 as 0 in the given array 
array
```

```python
array[array % 3 == 0] = 0
array[array % 5 == 0] = 0
print(array)
```
