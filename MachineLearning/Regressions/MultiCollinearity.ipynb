```python
import pandas as pd 
import seaborn as sns

import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv"
df = pd.read_csv(url)
df.head()

```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor # this is used to find the multicollinearity of the data 
vif = []

for i in range(3):
    vif.append(variance_inflation_factor(df.values, i))
```

```python
pd.DataFrame({'VIF': vif}, index=df.columns[1:4]).T # we don't need the last column because it is the target variable
```
