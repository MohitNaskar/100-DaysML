```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

def find_project_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p / ".git").exists() or (p / "Datasets").exists():
            return p
    raise FileNotFoundError(f"Project root not found from: {start}")

PROJECT_ROOT = find_project_root(Path.cwd())
DATASET_23 = PROJECT_ROOT / "Datasets" / "datasets-session-23"

batter_season = pd.read_csv(DATASET_23 / "batsman_season_record.csv")
```
