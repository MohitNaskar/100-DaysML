# Histogram


```python
tips = px.data.tips()
```

```python
tips.head()
```

```python
fig = px.histogram(tips, x='total_bill', color='smoker', title='Total Bill Distribution by Smoker Status')
fig.show()
```

# BoxPlot


```python
fig = px.box(tips, y= 'total_bill')
fig.show()
```
