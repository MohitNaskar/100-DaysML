```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 10 # number of trials
p = 0.5 # probability of success
size = 1000 # number of samples

bionomial = np.random.binomial(n, p, size)

# plt.hist(bionomial)
sns.distplot(bionomial)
plt.show()
```
