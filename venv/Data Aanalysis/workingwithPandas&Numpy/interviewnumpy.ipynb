# Plotting 3D graphs

```python
import numpy as np
```

```python
# meshgrids
#hink of it like turning two axes — x and y — into a full 2D grid of points, so you can compute something at each (x, y) location.
a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
```

```python
print(a.shape)
```

```python
xx,yy = np.meshgrid(a,b)
```

```python
yy.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(xx,yy)
plt.show()
```

```python
def function(x,y):
    return 3*np.log(x) + 2*x

zz = function(xx,yy)
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()

fig.add_trace(go.Surface(x=xx,y=yy,z=zz))

fig.show()
```
