There is an array of marks of 5 students in 4 subjects. Further you are asked to perform below task.

```python
marks = np.array([[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]])
marks
```
