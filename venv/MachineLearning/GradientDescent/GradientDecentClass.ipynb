```python
class GDRegressor:
    def __init__(self, n_iterations, learning_rate=0.01):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.m = 44.43
        self.b = -10

    def fit(self, X, y):
        for i in range(self.n_iterations):
            loss_slope = -2 * np.sum(y - self.m* X.ravel() - self.b)
            self.b = self.b - self.learning_rate * loss_slope

        print(self.b)
    def predict(self, X):
        return self.m * X.ravel() + self.b
```

```python
GDRegressor = GDRegressor(n_iterations=100, learning_rate=0.001)
GDRegressor.fit(X, y)
```
