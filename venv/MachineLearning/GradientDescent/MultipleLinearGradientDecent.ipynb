```python
class GradientDescent:
    def __init__(self, lr=0.01, num_iter=1000):
        self.lr = lr
        self.num_iter = num_iter
        self.coef_ = None
        self.intercept_ = None
        #x_train.shape will give us the number of coefficients we need to calculate


    def fit(self,X,y):
        self.intercept_ = 0
        self.coef_ = np.ones(X.shape[1])
        
        for i in range(self.num_iter):
            y_hat = self.intercept_ + np.dot(X, self.coef_)
            intercept_gradient = -2 * np.mean(y - y_hat)
            self.intercept_ = self.intercept_ - self.lr * intercept_gradient

            coef_gradient = -2 * np.dot((y - y_hat),X) / X.shape[0]
            self.coef_ = self.coef_ - self.lr * coef_gradient

        print(self.intercept_, self.coef_)

    def predict(self,X):
        return self.intercept_ + np.dot(X, self.coef_)
```

```python
gdr = GradientDescent(lr=0.01, num_iter=1000)
gdr.fit(X_train, y_train)
```
