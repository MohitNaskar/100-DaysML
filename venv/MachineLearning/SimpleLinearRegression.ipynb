```python
import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(lr.score(x_test,y_test))
```

```python
y_pred
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_test,y_pred,color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

# y = mx + b

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```

```python
# calculating a persons package who has cgpa of 7.5
cgpa = 7.5
package = m * cgpa + b
print(package) 
```
