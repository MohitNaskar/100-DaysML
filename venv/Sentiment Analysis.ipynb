```python
def analyze_sentiment(text):
    """
    Analyzes the sentiment of the given text using OpenAI's API
    Returns: sentiment (positive/negative/neutral) and explanation
    """
    
    prompt = f"""Analyze the sentiment of the following text and respond with ONLY one word: 
    'positive', 'negative', or 'neutral'.
    
    Text: {text}
    
    Sentiment:"""
    
    try:
        response = openai.chat.completions.create(
            model="gpt-3.5-turbo",  # Using the most affordable model
            messages=[
                {"role": "system", "content": "You are a sentiment analysis expert."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )
        
        sentiment = response.choices[0].message.content.strip().lower()
        return sentiment
    
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
if __name__ == "__main__":
    # Test texts
    texts = [
        "I absolutely love this product! It's amazing and works perfectly.",
        "This is the worst experience I've ever had. Completely disappointing.",
        "The weather today is normal. Nothing special.",
        "I'm so happy and excited about this opportunity!"
    ]
    
    print("=" * 50)
    print("SENTIMENT ANALYZER")
    print("=" * 50)
    
    for i, text in enumerate(texts, 1):
        print(f"\nText {i}: {text}")
        sentiment = analyze_sentiment(text)
        print(f"Sentiment: {sentiment.upper()}")
        print("-" * 50)
    
    # Interactive mode - analyze your own text
    print("\n\n" + "=" * 50)
    print("TRY IT YOURSELF!")
    print("=" * 50)
    
    user_text = input("\nEnter text to analyze (or press Enter to skip): ")
    
    if user_text.strip():
        sentiment = analyze_sentiment(user_text)
        print(f"\nYour text: {user_text}")
        print(f"Sentiment: {sentiment.upper()}")
    else:
        print("No input provided. Exiting...")
```
