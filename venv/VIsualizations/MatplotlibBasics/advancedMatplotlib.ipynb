# Subplots

```python
fig, ax = plt.subplots()
ax.scatter(batter['avg'], batter['strike_rate'])
ax.set_xlabel('Batting Average')
ax.set_ylabel('Strike Rate')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,sharex=True)
ax[0].scatter(batter['avg'], batter['strike_rate'])
ax[1].scatter(batter['avg'], batter['runs'])
ax[0].set_title('Batting Average vs Strike Rate')
ax[1].set_title('Batting Average vs Runs')
```

# Surface Plot
usually used for loss functions


```python
import numpy as np
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx,yy = np.meshgrid(x, y)
```

```python
z = xx**2 + yy**2
fig = plt.figure()
ax = plt.subplot(projection='3d')
a = ax.plot_surface(xx, yy, z,cmap='viridis')
fig.colorbar(a)
```
