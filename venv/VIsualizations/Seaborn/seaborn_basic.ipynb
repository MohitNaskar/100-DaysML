```python
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Scatter plot with seaborn
#scatter plot -> axes level
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
# figure level
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex',kind='scatter')
```

```python
# figure level
sns.relplot(data=tips, x='total_bill', y='tip', hue='day',kind='line')
```

```python
gap = px.data.gapminder()
```

```python
gap.head()
```

```python
gap[gap['country'] == 'India']
```

```python
# figure level
sns.relplot(data=gap, x='year', y='lifeExp',kind='line')
```

```python
gap2 = gap[gap['country'].isin(['India', 'China', 'Pakistan'])]
```

```python
# figure level
sns.relplot(data=gap2, x='year', y='lifeExp',hue='country',kind='line',row='country')
```

# KD plot

rather than using discreate bin, kde plot smooths the observations with a Gaussian kernel producing a continuous density eastimate

```python
sns.kdeplot(data=tips,x = 'total_bill',hue='day')
```

```python
# Rugplot

#Plot marginal distributions by drawing ticks along the x and y axes

# This function is intented to complement other plots by showing the location of individual observations in a unobtrusive way

sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram 
# a bivariate histogram bin the data witin rectangles that tile the plot
# and then shows the count of the observations within each reactangle with fill color 

sns.histplot(data=tips,x='total_bill',y='tip',cbar=True)
```

```python
sns.kdeplot(data=tips,x='total_bill',y='tip',fill=True)
```
